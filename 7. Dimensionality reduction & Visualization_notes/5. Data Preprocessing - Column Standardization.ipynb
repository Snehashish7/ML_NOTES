In <b><i>Column Normalization</i></b>, we took each column and compressed all the values between [0, 1] to get rid of scales of each feature. We created our own scale where all the values lie between 0 and 1. <b><i>aᵢ' ∈ [0,1]</i></b>

<i><u>Geometric interpretation</u> of Column Normalization</i>: The values are squished into an unit hypercube. 

<b><i>Column Standardization</i></b> is more often used in practice because it has a relationship with Gaussian distribution.

In column standardization, we standardize every column.

Take a column - n-values corresponding to the feature fⱼ = {a₁, a₂, ..., aᵢ, ... ,aₙ}. Take each column at a time and tansform it. - Process each d feature in X separately.

Each feature can be in any scale - kms, inches, etc. 

Transform all points in this column as: aᵢ' = {a₁', a₂', ..., aᵢ', ... ,aₙ'} - these points can be from any distribution. Convert the data points in such a way that the new points {a₁', a₂', ..., aᵢ', ... ,aₙ'}

<b><i> mean(aᵢ')ⁿᵢ₌₁ = 0 and standard deviation(aᵢ')ⁿᵢ₌₁ = 1</i></b>.


a̅ = mean{aᵢ}ⁿᵢ₌₁       ---> sample means

s = standard-deviation{aᵢ}ⁿᵢ₌₁       ---> sample standard deviation

<b><i> aᵢ' = (aᵢ - a̅) / s </i></b> --->guarantees mean(aᵢ')ⁿᵢ₌₁ = 0 and standard deviation(aᵢ')ⁿᵢ₌₁ = 1



<i><u>Geometric interpretation</u></i> 

Let us take a vector, with 2 features f₁(x-axis = heights(h)) and f₂(y-axis = weights(w)) in any scale.

Geometrically, if we have a bunch of points, mean vector is a sort of central datapoint or central vector.

x̄ = [hx̄, wx̄] where hx̄ = mean(hᵢ)ⁿᵢ₌₁ - central value on x-axis - mean height and wx̄ = mean(wᵢ)ⁿᵢ₌₁ - central value on y-axis - mean weight

Variance shows how spread the data is. The more the spread, the more the variance. 

By column standardization, I get 2 features f₁'(x-axis = heights(h')) and f₂'(y-axis = weights(w')). 



The new data will move to the origin(0,0) of the x-Y axis. The mean vector will lie at (0,0). 


<h5> <i><u>Mean Centering</u></i> Tranform the space such that the mean vector moves to (0,0) and squish or expand the data points such that the variance / standard deviation for any feature is 1. </h5>

- squish if standard dev > 1, expand if standard dev < 1

<h5> Column standardization is often called as mean centering where we are centering the mean at origin (0,0) and also scaling (ensuring that standard deviation on every axis is 1). </h5>
